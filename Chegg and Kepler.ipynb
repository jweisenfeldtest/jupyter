```python
#https://www.chegg.com/homework-help/questions-and-answers/kepler-s-law-orbital-motion-please-help-far-python-coding-showing-right-graph-could-tell-w-q40049782

import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint

y0=[0]
e = 0.3;
P = 1; #yr

t = np.linspace(0,1,100)

def model(u,t,e,P):
    theta = u
    return 2*np.pi*(1+e*np.cos(theta))**2/P/(1-e*e)**1.5

sol1 = odeint(model,y0,t,args=(0.0,1))
sol2 = odeint(model,y0,t,args=(0.3,1))
sol3 = odeint(model,y0,t,args=(0.6,1))
sol4 = odeint(model,y0,t,args=(0.9,1))

plt.plot(t, sol1, 'b', label='e=0.0')
plt.plot(t, sol2, 'y', label='e=0.3')
plt.plot(t, sol3, 'r', label='e=0.6')
plt.plot(t, sol4, 'g', label='e=0.9')
plt.legend(['e=0','e=0.3','e=0.6','e=0.9'])
plt.xlabel('t, yr')
plt.grid()
plt.show()
```
