```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
def model(u,t,G,M):
    r, theta, rdot, thetadot = u
    dydt = [rdot, thetadot, -G*M/r/r+r*thetadot*thetadot, 2*rdot*thetadot/r]
    return dydt
```

```python
y0 = [1,1,0.01,0.01]
G = 1 #6.67E-11 #Universal gravitation constant
M = 1 #2E30 #kg
```

```python
#import matplotlib.pyplot.bar as bar
#bar.plot(y0)
# https://matplotlib.org/3.1.1/gallery/statistics/barchart_demo.html#sphx-glr-gallery-statistics-barchart-demo-py
```

```python
plt.plot(y0)
plt.title('initial conditions')
#plt.legend('legend')
plt.xlabel('r theta rdot thetadot')
plt.show()
# https://www.sitepoint.com/plot-charts-python-matplotlib/
```

```python
t=np.linspace(0,100,1001)
```

```python
from scipy.integrate import odeint
sol = odeint(model,y0,t, args=(G, M))
```

```python
import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
plt.plot(t,sol[:,0],'b',label='r(t)')
plt.title('r(t)')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
plt.plot(t,sol[:,1],'b')
plt.title('theta(t)')
plt.xlabel('t')
plt.grid()
plt.show()
```
